```python
import numpy as np
```

```python
def fitness_calculation(individual):
    """Calculate fitness for individual"""
    x = individual
    result = (x[3]**1)*(x[4]**1) + x[2]*x[3]*x[4]**2 + x[3] + x[1]*x[3] + \
             (x[1]**2)*(x[2]**1)*(x[3]**2)*x[4] + 50
    return abs(result)
```

```python
def make_individual(number_of_genes, lower_limit, upper_limit):
    """Make individual"""
    individual = np.random.randint(lower_limit, upper_limit, number_of_genes)
    return individual
```

```python
def make_first_population(number_of_individuals, number_of_genes = 5, lower_limit = -100, upper_limit = 100):
    """Make first population"""
    population = [make_individual(number_of_genes, lower_limit, upper_limit) for _ in range(number_of_individuals)]
    return population
```

```python
def selection(population):
    survived_individuals = []
    np.random.shuffle(population)
    for i in range(0,len(population)-1,2):
        if abs(fitness_calculation(population[i])) < abs(fitness_calculation(population[i+1])):
            survived_individuals.append(population[i])
        else:
            survived_individuals.append(population[i+1])
    return survived_individuals
```

```python
def pairing(selected):
    parents = []
    for i in range(2):
        np.random.shuffle(selected)
        for i in range(0, len(selected)-1,2):
            parents.append([selected[i],selected[i+1]])
    return parents
```

```python
def single_point_mating(parents, point=3,p1=0.1,p2=0.1):
    children = []
    children.append(np.array(list(parents[0][:point])+list(parents[1][point:])))
    children.append(np.array(list(parents[1][:point])+list(parents[0][point:])))
    return children
```

```python
def two_points_mating(parents, point_1 = 2, point_2 = 4):
    children = []
    children.append(np.array(list(parents[0][:point_1])+list(parents[1][point_1:point_2])+list(parents[0][point_2:])))
    children.append(np.array(list(parents[1][:point_1])+list(parents[0][point_1:point_2])+list(parents[1][point_2:])))
    return children
```

```python
def mutation(child, p1 = 0.10,p2 = 0.2):
    new_child = []
    for i in child:
        if np.random.uniform() <= p1:
            new_child.append(np.random.randint(-300,300))
        elif np.random.uniform() <= p2:
            new_child.append(np.random.randint(-300,300))
        else:
            new_child.append(i)
    return np.array(new_child)
```

```python
population = make_first_population(10,5,-300,300)
for i in range(100):
    selected_population = selection(population)
    parents = pairing(selected_population)
    children = []
    for pair in parents:
        single_child = single_point_mating(pair)
        multi_child = two_points_mating(pair)
        children.append(single_child[0])
        children.append(single_child[1])
        children.append(multi_child[0])
        children.append(multi_child[1])
    new_children = []
    for child in children:
        new_children.append(mutation(child))
    arr = []
    for child in new_children:
        arr.append(fitness_calculation(child))
    arr = np.array(arr)
    new_children = np.array(new_children)
    arr1inds = arr.argsort()
    sorted_arr = arr[arr1inds[::1]]
    sorted_new_children = new_children[arr1inds[::1]]
    new_chil = [np.array(c) for c in sorted_new_children]
    population = new_chil[:10]
    print('----------------------------')
    for p in population:
        print(p)
```
